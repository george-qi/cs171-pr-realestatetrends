```python
import csv
import json
```

```python
def init_data(fname):
    csv_filename = fname +'.csv'

    with open(csv_filename, 'rU') as csv_fh:
        # Parse as a CSV file.
        reader = csv.reader(csv_fh)

        # Skip the header line.
        header = next(reader, None)

        # Loop over the file.
        data = []

        for row in reader:
            item = {
                'month': row[0],
                'nasdaq_volume': row[5],
                'nasdaq_adj_close': row[6],
                'nasdaq_perc': row[7],
                'spy_volume': -1,
                'spy_adj_close': -1,
                'spy_perc': -1
            }
            data.append(item)

    return data 
```

```python
init_data("nasdaq")
```
